```python
# Calculate the size of each chunk
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)

#Split the dataframe into chunks and save them
for i in range(5):
    start = i * chunk_size
    end = start + chunk_size
    chunk_df = df.iloc[start:end]
    #chunk_df.to_csv(f"/Users/17084/Desktop/Ecom_Azure_E2E/data/chunk_data/chunk{i+1}.csv", index = False)
    chunk_df.to_csv(f"chunk_data/chunk{i+5+1}.csv")
```
